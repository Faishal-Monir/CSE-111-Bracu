```python
#dic+tuple task-
#a: 100, b: 100, c: 200, d: 300
#a: 300, b: 200, d: 400, e: 200


#dic+tuple task-1
data_a=input("enter dic-1:")
data_b=input("enter dic-2:")
data_a=data_a.split(", ")
data_b=data_b.split(", ")

dic={}
dic_2={}
for i in data_a:
    x,y=i.split(": ")
    dic[x]=int(y)
    
for j in data_b:
    m,n=j.split(": ")
    dic_2[m]=int(n)

output={}
for k in dic.items():
    a,b=k
    output[a]=b
for l in dic_2.items():
    c,d=l
    if c not in output.keys():
        output[c]=d
    else:
        d=dic[c]+dic_2[c]
        output[c]=d
lst=[]
for p in output.values():
    if p not in lst:
        lst.append(p)
lst.sort()
lst=tuple(lst)
print(output)
print("Values:",lst)
```

```python
#dic+tuple task-3   #Input:  key1 : value1, key2 : value2, key3 : value1
data=input("enter your dictionary:").split(", ")
store={}
for i in data:
    a,b=i.split(" : ")
    store[a]=b

output={}
for j in store.items():
    c,d=j
    if d not in output.keys():
        output[d]=[c]
    else:
        output[d].append(c)
print(output)
```

```python
#functiontask-1

def BMI(height,weight):
    lenght=height/100
    calc=weight/(lenght**2)
    if calc< 18.5: 
        return "Score is {0:3.1f}. You are Underweight".format(calc)
    elif 18.5<= calc<=24.9:
        return "Score is {0:3.1f}. You are Normal".format(calc)
    elif 25<= calc<=30:
        return "Score is {0:3.1f}. You are Overweight".format(calc)
    elif calc>30:
        return "Score is {0:3.1f}. You are Obese".format(calc)
x=BMI(152,48)
print(x)
```

```python
#functiontask-2
def onko(minimum, maximum,divisor):
    
    store=0
    for i in range(minimum,maximum):
        if i%divisor==0:
            store+=i
        else:
            pass
    return store    
onko(3,16,3)
```

```python
#functiontask-4
def replace_domain(test,new,old="kaaj.com"):
    
    store=""
    if new in test:
        print("Unchanged:",test)
    else:
        for i in test:
            if i!="@":
                store+=i
            elif i=="@":
                store+=i
                break
        store+=new
        print("Changed:",store)   
replace_domain('alice@kaaj.com','sheba.xyz','kaaj.com')    
```

```python
#functiontask-5
def palindrome(x):
    store=""
    
    for i in x:
        if i==" ":
            pass
        else:
            store+=i
    rev=store[-1::-1]
    
    if store==rev:
        print("Palindrome")
    elif store!=rev:
        print("Not a palindrome")
    
palindrome("madam")
```
